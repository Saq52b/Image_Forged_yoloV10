```python
import requests
import cv2
import numpy as np
import imutils
import matplotlib.pyplot as plt
import supervision as sv
from ultralytics import YOLOv10
import keyboard

# Initialize the YOLOv10 model
model = YOLOv10('./runs/detect/train3/weights/last.pt')

# Replace the below URL with your own. Make sure to add "/video" for live streaming.
url = "http://192.168.0.89:8080/video"

# Capture video stream from the URL
cap = cv2.VideoCapture(url)

# Create a single figure and axis for plotting
plt.ion()
fig, ax = plt.subplots()
ax.set_title("Android_cam")
ax.axis('off')

while True:
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame")
        break

    frame = imutils.resize(frame, width=1000, height=1800)

    # Run YOLOv10 model on the captured frame
    results = model(frame)[0]
    detections = sv.Detections.from_ultralytics(results)
    bounding_box_annotator = sv.BoundingBoxAnnotator()
    label_annotator = sv.LabelAnnotator()

    # Annotate the frame with bounding boxes and labels
    annotated_frame = bounding_box_annotator.annotate(scene=frame, detections=detections)
    annotated_frame = label_annotator.annotate(scene=annotated_frame, detections=detections)

    # Convert the frame to RGB (matplotlib uses RGB by default)
    annotated_frame_rgb = cv2.cvtColor(annotated_frame, cv2.COLOR_BGR2RGB)

    # Update the same plot
    ax.imshow(annotated_frame_rgb)
    plt.draw()
    plt.pause(0.001)
    ax.clear()
    ax.set_title("Android_cam")
    ax.axis('off')

    # Press 'Esc' key to exit
    if keyboard.is_pressed('esc'):
        break

# Release the capture and destroy any OpenCV windows
cap.release()
cv2.destroyAllWindows()
plt.ioff()
plt.show()
```
